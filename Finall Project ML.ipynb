```python
# Split the data into features (X) and target (y)
x = data[['number of bedrooms', 'number of bathrooms',
       'living area','condition of the house', 'Number of schools nearby']]#features(5)
y = data[['Price']]#Target(1)
```

```python
# Splitting the Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
# test_size=0.2 → 20% of the data will be used for testing, and 80% for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Return the number of rows and columns in the DataFrame as a tuple (rows, columns)
x_train.shape
```

```python
# Import the GridSearchCV tool for hyperparameter tuning.
from sklearn.model_selection import GridSearchCV
```

```python
# Decision Tree Regression is a supervised learning
#predicts continuous values using a tree structure.
# It splits data into subsets based on feature values.

# Nodes:
# - Root: whole dataset
# - Decision: splits based on features
# - Leaf: predicted value

# How it works:
# 1. Find the best feature to split the data.
# 2. Split data and repeat for each subset.
# 3. Stop when max depth or min samples reached.

# Prediction:
# - Follow splits to a leaf node; leaf value is the prediction.

# Pros: handles non-linear data, easy to visualize.
# Cons: can overfit, sensitive to data changes.
```

```python
# Import the Decision Tree Regressor model.
from sklearn.tree import DecisionTreeRegressor
# Define the grid of hyperparameters to search through.
param_grid = {
"criterion" : ["mse", "friedman_mse", "mae"], # Criteria for measuring the quality of a split (e.g., 'mse' or 'mae').
"splitter" : ["best", "random"],# Strategy used to choose the split point ('best' or 'random').
"max_depth" : [None, 10, 20, 30 , 40, 50],# Maximum depth of the tree (None means unlimited depth).
"min_samples_split" : [2,5,10],# Minimum number of samples required to split an internal node.
"min_samples_leaf" : [1,2,4]# Minimum number of samples required to be at a leaf node.
}
```

```python
# Initialize the Decision Tree Regressor model.
tree_model = DecisionTreeRegressor()
```

```python
# Initialize GridSearchCV:
# estimator: The model to tune (DecisionTreeRegressor).
# param_grid: The dictionary containing hyperparameter values to test.
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
# Train the model using the training data
# X_train: features (independent variables)
# y_train: target (dependent variable)
# The model learns the relationship between X_train and y_train.
grid_tree.fit(x_train, y_train)
```

```python
# Display the best hyperparameters found by the Grid Search process
# These parameters will be used in the final, optimized Decision Tree model
grid_tree.best_params_
```

```python
# Use the best model found by GridSearchCV to make predictions on the test set
y_pred = grid_tree.predict(x_test)
```

```python
# Import necessary evaluation metrics from scikit-learn
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics

# Mean Absolute Error (MAE): The average magnitude of the errors (measures average error in the same units as the target variable)
mae = mean_absolute_error(y_test, y_pred)
# Mean Squared Error (MSE): The average of the squared errors  (measures the average squared difference between actual values (y_test) and predicted values (y_pred))
# Formula: MSE = (1/n) * Σ(y_i - y_pred_i)^2
# Lower MSE indicates better predictions
mse = mean_squared_error(y_test, y_pred)
# Root Mean Squared Error (RMSE): The square root of the MSE (interpretable in the same units as the target variable)
rmse = np.sqrt(mse)
# R-squared (R2): The coefficient of determination (measures thehow well the model explains the variance in the target)
# 1 → perfect prediction, 0 → model predicts as well as the mean, negative → worse than mean
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:,.2f}")
print(f"Mean Squared Error (MSE): {mse:,.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
# Linear Regression Explanation

# Linear Regression is a supervised machine learning algorithm
# used to model the relationship between a dependent variable (target)
# and one or more independent variables (features).

# The main idea is to find the best-fit straight line
# that minimizes the difference between the predicted values
# and the actual values in the dataset.

# This line is represented by the equation:
# y = b0 + b1*x
# where:
# y  = predicted value
# b0 = intercept (value of y when x = 0)
# b1 = slope (amount of change in y for a one-unit change in x)
# x  = input feature

# The algorithm works by minimizing an error function,
# usually Mean Squared Error (MSE),
# which measures how far predictions are from actual values.

# Linear Regression assumes:
# 1. A linear relationship between input and output variables
# 2. Independence of observations
# 3. Constant variance of errors (homoscedasticity)
# 4. Normally distributed errors
# 5. No strong multicollinearity between features

# It is commonly used for:
# - Price prediction
# - Trend analysis
# - Forecasting numeric values

# Linear Regression is simple, fast, and easy to interpret,
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
```

```python
# Create an instance of the Linear Regression model
# This initializes the model with default parameters.
lin_model = LinearRegression()
```

```python
# Train the model using the training data
# X_train: features (independent variables)
# y_train: target (dependent variable)
# The model learns the relationship between X_train and y_train.
lin_model.fit(x_train, y_train)

# Predict target values for the test set
# X_test: features of unseen data
# y_pred: predicted values based on the learned relationship
y_pred_lin = lin_model.predict(x_test)
```

```python
# Mean Absolute Error (MAE): The average magnitude of the errors (measures average error in the same units as the target variable)
# Mean Squared Error (MSE): The average of the squared errors  (measures the average squared difference between actual values (y_test) and predicted values (y_pred))
# Formula: MSE = (1/n) * Σ(y_i - y_pred_i)^2
# Lower MSE indicates better predictions
# Root Mean Squared Error (RMSE): The square root of the MSE (interpretable in the same units as the target variable)
# R-squared (R2): The coefficient of determination (measures thehow well the model explains the variance in the target)
# 1 → perfect prediction, 0 → model predicts as well as the mean, negative → worse than mean
```

```python
#Evaluation
print("Linear Regression Results:")
print("MSE =", mean_squared_error(y_test, y_pred_lin))
print("MAE =",mean_absolute_error(y_test,y_pred_lin ))
print("R2 Score =", r2_score(y_test,y_pred_lin))
```

```python
# Random Forest Regressor Explanation

# Random Forest Regressor is a supervised learning algorithm
# used to predict continuous numeric values.

# It is based on building multiple decision trees
# using random samples of data and features.

# Each tree makes a prediction,
# and the final result is the average of all trees.

# This approach reduces overfitting
# and improves prediction accuracy.

# Random Forest can model non-linear relationships
# and handle complex data patterns.

# It does not require feature scaling
# but is less interpretable and more computationally expensive
# than simple models like Linear Regression.
```

```python
# انسخ كودك كاملاً كما هو هنا
%%writefile app_exact.py
import streamlit as st
import joblib
import numpy as np
import pandas as pd
import json
import os

# =================================================================
# 1. Page Configuration (Must be the first command)
# =================================================================
st.set_page_config(
    page_title="House Price Predictor ",
    layout="wide", # Use wide layout to maximize space
    initial_sidebar_state="collapsed",
)

# =================================================================
# 2. Load Model and Features Safely
# =================================================================
model_path = 'house_price_model.pkl'
feature_info_path = 'model_features.json'

# Initialize FEATURE_NAMES with a default list to prevent NameError
FEATURE_NAMES = [
    'number of bedrooms',
    'number of bathrooms',
    'living area',
    'condition of the house',
    'number of floors',
    'Number of schools nearby'
]

model = None # Initialize model

if os.path.exists(model_path):
    # Load the trained model (e.g., the best Random Forest Regressor)
    model = joblib.load(model_path)

    # Load feature names if available, otherwise use default
    if os.path.exists(feature_info_path):
        with open(feature_info_path, 'r') as f:
            feature_info = json.load(f)
            FEATURE_NAMES = feature_info['feature_names']
    else:
        st.warning("Feature names file (model_features.json) not found. Using default feature names for prediction.")
else:
    st.error(" Model file (house_price_model.pkl) not found. Please train and save the model first.")
    st.info("To train and save the model, please execute the relevant training cells in this notebook before running this Streamlit app code.")
    st.stop() # This stops the Streamlit app execution if the model is not found

# =================================================================
# 3. User Interface Design (UI Design)
# =================================================================

st.title(" House Price Prediction App")
st.markdown("---")

st.subheader("Enter Property Features")
st.markdown("Please enter the house characteristics to get a price estimate from the Machine Learning model.")


# Use three columns for better organization
col1, col2, col3 = st.columns(3)

with col1:
    # Use number_input for discrete/float values
    bedrooms = st.number_input("🛏 Number of Bedrooms", min_value=1, max_value=10, value=3, step=1)
    bathrooms = st.number_input(" Number of Bathrooms (0.5 represents a half-bath)", min_value=0.5, max_value=10.0, value=2.5, step=0.5)

with col2:
    living_area = st.number_input(" Living Area (sq ft)", min_value=100, value=2000, step=100)

    # Use st.slider for a better user experience for categorical/ordinal features
    condition = st.slider(" House Condition (1 = Poor, 5 = Excellent)", min_value=1, max_value=5, value=3, step=1)

with col3:
    floors = st.slider("Number of Floors", min_value=1.0, max_value=5.0, value=2.0, step=0.5)
    schools = st.number_input(" Number of Schools Nearby", min_value=0, max_value=10, value=5, step=1)

st.markdown("---")


# =================================================================
# 4. Prediction Logic and Result Display
# =================================================================

# Prepare features in the exact order matching the model
feature_values = []

# This loop ensures the correct order and fills in non-UI-controlled features (like OHE) with 0
for feature_name in FEATURE_NAMES:
    if feature_name == 'number of bedrooms':
        feature_values.append(bedrooms)
    elif feature_name == 'number of bathrooms':
        feature_values.append(bathrooms)
    elif feature_name == 'living area':
        feature_values.append(living_area)
    elif feature_name == 'condition of the house':
        feature_values.append(condition)
    elif feature_name == 'number of floors':
        feature_values.append(floors)
    elif feature_name == 'Number of schools nearby':
        feature_values.append(schools)
    else:
        # Default value of 0 for any features not explicitly controlled by UI (e.g., unselected OHE columns)
        feature_values.append(0)

# Create DataFrame with exact feature names and single row of inputs
feature_df = pd.DataFrame([feature_values], columns=FEATURE_NAMES)

# Prediction button

if st.button(" Predict Price", type="primary", use_container_width=True):
    try:
        # Make prediction
        prediction = model.predict(feature_df)[0]

        # Display Success Message and animation
        st.success(" Prediction Successful!")
        st.toast(' Price predicted successfully!', icon='🏠')

        st.markdown("---")

        # Use st.metric for a prominent and attractive display of the result
        st.metric(
            label=" Predicted House Price (in currency units)",
            value=f"${prediction:,.2f}",
            delta="Machine Learning Model Estimate", # Descriptive text
            delta_color="normal"
        )

        # Optional: Display input summary for confirmation
        with st.expander(" View Input Summary"):
            st.dataframe(feature_df)

    except Exception as e:
        st.error(f" Error during prediction process: {str(e)}")
        st.warning("Please ensure all inputs are valid and the model is loaded correctly.")
else:
    st.info(" Enter house features above and click the button to get the prediction")

st.markdown("---")

# =================================================================
# 5. Enhanced Footer (Styling for the development team)
# =================================================================
st.markdown("""
    <style>
    /* Improve Primary Button Style */
    .stButton>button {
        font-size: 1.2rem;
        font-weight: bold;
    }
    /* Enhance the look of the predicted value in st.metric */
    div[data-testid="stMetricValue"] {
        font-size: 3rem;
        font-weight: bolder;
        color: #764ba2;
    }
    </style>

    <div style='text-align: center; padding: 20px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); border-radius: 10px; margin-top: 30px;'>
        <h3 style='color: white; margin-bottom: 15px;'>👥 Developed By</h3>
        <div style='display: flex; justify-content: center; flex-wrap: wrap; gap: 15px; color: white;'>
            <div style='background: rgba(255,255,255,0.2); padding: 10px 15px; border-radius: 8px;'>✨ Rana Fayez 235324</div>
            <div style='background: rgba(255,255,255,0.2); padding: 10px 15px; border-radius: 8px;'>✨ Judy Yasser 235291</div>
            <div style='background: rgba(255,255,255,0.2); padding: 10px 15px; border-radius: 8px;'>✨ Salma Ashraf 235383</div>
            <div style='background: rgba(255,255,255,0.2); padding: 10px 15px; border-radius: 8px;'>✨ Malak Emad 235287</div>
            <div style='background: rgba(255,255,255,0.2); padding: 10px 15px; border-radius: 8px;'>✨ Menna Mohamed 235229</div>
        </div>
        <p style='color: rgba(255,255,255,0.9); margin-top: 15px; font-size: 14px;'>🏠 House Price Prediction using Machine Learning</p>
    </div>
    """, unsafe_allow_html=True)

print("✅ the code is saved app_exact.py")
```
